```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient =  df(previous_x) ## [REQUIRED]
    new_x = previous_x - step_multiplier * gradient ## [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)
    print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))## [REQUIRED])

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

# Example 4 - Data Viz with 3D Charts

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples
# Define the function f(x, y)
def f(x, y):
    return 1 / (3**(-x**2 - y**2) + 1)

# Define the gradient of f(x, y)
def gradient_f(x, y):
    exp_term = 3**(-x**2 - y**2)
    df_dx = (2 * x * exp_term * np.log(3)) / (exp_term + 1)**2
    df_dy = (2 * y * exp_term * np.log(3)) / (exp_term + 1)**2
    return np.array([df_dx, df_dy])
def gradient_descent_2d(initial_guess, learning_rate=0.01, precision=0.00001, max_iter=500):
    new_x, new_y = initial_guess
    x_list = [new_x]
    y_list = [new_y]
    cost_list = [f(new_x, new_y)]

    for _ in range(max_iter):
        previous_x, previous_y = new_x, new_y
        gradient = gradient_f(previous_x, previous_y)
        new_x -= learning_rate * gradient[0]
        new_y -= learning_rate * gradient[1]

        x_list.append(new_x)
        y_list.append(new_y)
        cost_list.append(f(new_x, new_y))

        step_size = np.sqrt((new_x - previous_x)**2 + (new_y - previous_y)**2)
        if step_size < precision:
            break

    return (new_x, new_y), x_list, y_list, cost_list
# Run gradient descent with different learning rates
n = 100
initial_guess = (0.5, 0.5)

# Low learning rate
low_lr_result = gradient_descent_2d(initial_guess, learning_rate=0.001, max_iter=n)

# Mid learning rate
mid_lr_result = gradient_descent_2d(initial_guess, learning_rate=0.01, max_iter=n)

# High learning rate
high_lr_result = gradient_descent_2d(initial_guess, learning_rate=0.1, max_iter=n)

# Plot the results
plt.figure(figsize=[15, 10])

# Low learning rate
plt.subplot(2, 2, 1)
plt.title('Cost Reduction - Low Learning Rate (0.001)', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(range(len(low_lr_result[3])), low_lr_result[3], color='blue', linewidth=2, alpha=0.8)

# Mid learning rate
plt.subplot(2, 2, 2)
plt.title('Cost Reduction - Mid Learning Rate (0.01)', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(range(len(mid_lr_result[3])), mid_lr_result[3], color='blue', linewidth=2, alpha=0.8)

# High learning rate
plt.subplot(2, 2, 3)
plt.title('Cost Reduction - High Learning Rate (0.1)', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(range(len(high_lr_result[3])), high_lr_result[3], color='blue', linewidth=2, alpha=0.8)

plt.tight_layout()
plt.show()

# Print the results
print('Low Learning Rate (0.001) - Final point:', low_lr_result[0])
print('Mid Learning Rate (0.01) - Final point:', mid_lr_result[0])
print('High Learning Rate (0.1) - Final point:', high_lr_result[0])
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
